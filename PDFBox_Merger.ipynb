```python
clean_1 = result_df.dropna()
display(clean_1)
```

```python
clean_2 = clean_1.copy(deep=True)

# Define a function to map values
def bug_flagger(bugs):
    if bugs > 0:
        return 1
    else:
        return 0

# Apply the function to create the 'bug' column
clean_2['bug'] = clean_2['Bugs'].apply(lambda x: bug_flagger(x))

display(clean_2)
```

```python
clean_3 = clean_2.copy(deep=True)
clean_3 = clean_3.drop(['Class', 'Bugs', 'Version'], axis=1)
clean_3['temp'] = 1
display(clean_3)
```

```python
PDFBox_v31 = pd.read_csv("/content/PDFBox_MR/pdfBox-2.0.31.csv", header=1)
PDFBox_v31_clean = PDFBox_v31.dropna(how='any', ignore_index='True')
PDFBox_v31_clean['Version'] = '2.0.31'
PDFBox_v31_clean = PDFBox_v31_clean.drop(['CBO', 'NOC'],axis=1)
# Apply the mapping function to the 'Class' column
PDFBox_v31_clean['Class'] = PDFBox_v31_clean['Class'].apply(map_class)
display(PDFBox_v31_clean)
```

```python
with open('/content/Compute_PDFBox/PDFBox_2.0.31_compute.json', 'r') as file:
    data = json.load(file)

# Convert dictionary to Pandas DataFrame
Box_v31 = pd.DataFrame(data.items(), columns=['Class', 'Bugs'])
display(Box_v31)
```

```python
targets = pd.merge(Box_v31, PDFBox_v31_clean, on='Class', how='inner')
targets = targets.dropna()
# .apply(lambda x: bug_flagger(x)).drop(['Class', 'Bugs', 'Version'], axis=1)
targets['bug'] = targets['Bugs'].apply(lambda x: bug_flagger(x))
targets['temp'] = 0
targets = targets.drop(['Class', 'Bugs', 'Version'], axis=1)
display(targets)
```

###**Model Training.**

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
X_train = clean_3.drop('bug', axis=1)
y_train = clean_3['bug']
X_test = targets.drop("bug", axis=1)
y_test = targets["bug"]
```

```python
display(X_train)
display(y_train)
display(X_test)
display(y_test)
```

```python
# perform grid search to optimize the tree's depth
param_grid = {"max_depth": range(1, 11)}
decision_tree = DecisionTreeClassifier(random_state=0)
grid_search = GridSearchCV(decision_tree, param_grid=param_grid, cv=10)
grid_search.fit(X_train, y_train)

# create and train a decision tree model with the best depth found by grid search
decision_tree = grid_search.best_estimator_

# evaluate the model's performance using 10-fold cross-validation
cv_scores = cross_val_score(decision_tree, X_train, y_train, cv=10)
print("Cross-validation scores:", cv_scores)
print("Mean cross-validation score: {:.2f}%".format(cv_scores.mean() * 100))

# make predictions on the test data
y_pred = decision_tree.predict(X_test)

# evaluate the model's performance on the test data
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)

print("Accuracy: {:.2f}%".format(accuracy * 100))
print("Precision: {:.2f}%".format(precision * 100))
print("Recall: {:.2f}%".format(recall * 100))
print("F1 Score: {:.2f}%".format(f1 * 100))
print("Confusion Matrix:\n", confusion)

```
